```python
import pandas as pd
pd.options.display.max_rows = 10

# DND MODE
import warnings
warnings.filterwarnings("ignore")
```

```python
# the column names for each table is in the README.md
# Based on that lets import our tables to df
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('Data/users.dat',sep='::',header=None,names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('Data/ratings.dat', sep='::',header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('Data/movies.dat', sep='::',header=None, names=mnames)
```

```python
# Lets check the tables
```

```python
users.head()
```

```python
movies.head()
```

```python
ratings.head()
```

```python
# lets join them for easier analysis
data = pd.merge(pd.merge(ratings,users),movies)
```

```python
data.head()
```

```python
# get the mean movie rating for each film grouped by gender
data.groupby(['title','gender'])['rating'].mean().unstack()
```

```python
# the same thing can be accomplished using pivot_table like this

data.head()
```

```python
mean_ratings = data.pivot_table('rating',index='title',columns='gender')
```

```python
mean_ratings.head()
```

```python
# now lets select the movies with atleast 250 ratings
```

```python
ratings_by_title = data.groupby(['title']).size()
```

```python
ratings_by_title>=250
```

```python
active_titles = ratings_by_title[ratings_by_title>=250].index
```

```python
mean_ratings = mean_ratings.loc[active_titles]
```

```python
mean_ratings
```

```python
# TOP FILMS AMONG FEMALE VIEWERS
top_female_ratings = mean_ratings.sort_values(by='F',ascending=False)
```

```python
top_female_ratings
```

```python
# lets see male ones
top_male_ratings = mean_ratings.sort_values(by='M',ascending=False)
```
